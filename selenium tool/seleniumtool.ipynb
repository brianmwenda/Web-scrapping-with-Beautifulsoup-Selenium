```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
import pandas as pd
import csv
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python


url = "https://www.covid19india.org/"
driver = webdriver.Chrome(executable_path='E:\FREELANCE\stano ombork\data science\data scrapping tool\selenium tool\chromedriver')
driver.get(url)
sleep(2)
  
def extractor():
    TCases = driver.find_element_by_xpath('/html/body/div/div/div[3]/div[1]/div[2]/div[1]/div[1]')
    TActive = driver.find_element_by_xpath('/html/body/div/div/div[3]/div[1]/div[2]/div[1]/div[2]')
    TRecov = driver.find_element_by_xpath('/html/body/div/div/div[3]/div[1]/div[2]/div[1]/div[3]')
    TDeath = driver.find_element_by_xpath('/html/body/div/div/div[3]/div[1]/div[2]/div[1]/div[4]')
    
   
    
    fields = ['Country', 'Cases', 'Deaths', 'Region'] 
    with open('covid .csv', 'w') as f:

        # using csv.writer method from CSV package
        write = csv.writer(f)

        write.writerow(TCases.text)
        write.writerows(TActive.text)
        write.writerows(TRecov.text)
        write.writerows(TDeath.text)
        
    driver.close()
    
    
    
    dataset = pd.read_csv("covid.csv",encoding='latin-1')
    sns.relplot(x="Cases", y="Deaths", data=dataset);
    sns.lmplot(x="Cases", y="Deaths", data=dataset);
while True:
    extractor()




    
```
